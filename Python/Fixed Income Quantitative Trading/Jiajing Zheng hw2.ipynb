PCA:
We decompose the covariance matrix of the returns into eigenvectors and eigenvalues such that:

$$\Sigma = E\Lambda E^T $$

The matrix of eigenvectors $E$ describes how much each original maturity/instrument (we will use maturity/instrument interchangeably) contributes to each factor, and $\Lambda$, a diagonal matrix of corresponding eigenvalues, describes how much of the overall daily return move is explained by each factor. I prefer to think of this in geometric terms. We are projecting from a universe of 9 instruments into a universe of 3 factors. These 3 factors are chosen to explain a majority of the move. 

This is essentially a $PDP^{-1}$ decomposition of the covariance matrix. Therefore, $E^{-1} = E^T$.

The matrix $E$ can also be called the factor loadings. We care about the first 3 columns of $E$, which corresponding to the first 3 factors (assuming that the matrix has been sorted in order of highest corresponding eigenvector. The dimension of the first 3 columns is $n\times 3$, where n is the number of instruments we have. Every day we have $n \times 1$ returns, which we will denote by $dy$. Every day, we can get the current level of the factors by doing $F = E^T\times dy$. The dimension of $F$ will be $3 \times 1$, which is what we want.

### i.	Use Chou-Ng estimation procedure or Box-Tiao

```python
import statsmodels.api as sm
def Box_Tiao(df):
    df = df-df.mean()
    df_lag = df.shift().dropna()
    df1 = df.loc[df_lag.index, :]
    model = sm.OLS(df1.values, df_lag.values).fit()
    A = model.params
    cov = df1.cov()
    Q = np.linalg.inv(cov)@ (np.transpose(A)) @ (cov) @ (A)
    val, vec = np.linalg.eig(Q)
    return val, vec
```

```python
val, vec = Box_Tiao(sample_df[["3 YR","5 YR","10 YR"]]) # vec: the column v[:,i] is the eigenvector corresponding to the eigenvalue w[i]
val, vec
```

```python
def halflife(x):
    x_lag = x.shift(1)
    x_lag.iloc[0] = x_lag.iloc[1]
    x_ret = x - x_lag
    x_ret.iloc[0] = x_ret.iloc[1]

    x_lag2 = sm.add_constant(x_lag)
    model = sm.OLS(x_ret,x_lag2)
    res = model.fit()

    halflife = round(-np.log(2) / list(res.params)[1], 4)
    return halflife
```

What is CCA? Find 2-3 examples

CCA is a way of inferring information from cross-covariance matrices.  If we have two vectors X,Y, and there are correlation among the variables, then CCA analysis will find linear combination of X and Y that have maximum correlation with each other. (Wikipedia)

A typical use of cca is to take two sets of variables and see what is the common among them. For example, in pyshological testing, one could take two well established multidimensional personality tests such as Minnesota Multiphasic Personality Inventory and NEO. By seeing how the MMPI factos relate to the NEO factors, one can understand what dimensions were common between these two tests and see how much variance was shared. 

People can also use cca to generate a model equation relating two sets of variables. We could investigate a set of performance measure and a set of explanatory variables. Constraint restrictions will be put to ensure that the model reflects theorectical requirement. This model is known as maximum correlation model. A researcher has collected data on three psychological variables, have academic variables, and gender for 500 college students. The researcher can use cca to investigate how many canonical variables are necessary to understand the association between the two sets of variables.
